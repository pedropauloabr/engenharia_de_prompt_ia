A unidade 2 da matéria engenharia de prompt e aplicações em ia desenvolveu e nos mostrou o uso correto da IA para um programador, não como um substituto mas sim um ótimo AUXILIAR.
Aprendendo o uso da forma correta para aprender ao mesmo tempo que realizamos o trabalho, sempre conferindo e corrigindo erros.
